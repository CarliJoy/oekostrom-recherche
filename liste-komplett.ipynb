```python
import pandas as pd
import time
```

# Komplette Liste
Zuerst füllen wir die Anbieter aus, die keine 100%-gen Ökostromanbieter sind. 
Diese scheitern bereits an RoWo Kiterium A. 

```python
merged = pd.read_csv("lichtblick/Lichtblick-RoWo-merged-20190826-cleaned.csv")
merged[['Firmenname', 'RoWo Liste', 'Erneuerbare Energien 2','RoWo-Kriterien', 'A']].head()
```

```python
print(len(merged.loc[merged['RoWo-Kriterien'] == 'A']))

merged.loc[merged['Erneuerbare Energien 2'] != "100.00%", 'RoWo-Kriterien'] = 'A'
merged.loc[merged['Erneuerbare Energien 2'] != "100.00%", 'A'] = 'x'
merged.loc[merged['Erneuerbare Energien 2'] != "100.00%", 'RoWo Liste'] = True

print(len(merged.loc[merged['RoWo Liste'] != True]))
```

Dann wird germergt!

```python
oeko_list = pd.read_csv("merged_data/oeko_list_large-20190826-cleaned.csv")
oeko_list

list_ = pd.merge(merged,
               oeko_list,
               on="Firmenname",
               how="outer")

print(merged.shape)
print(oeko_list.shape)
print(list_.shape)

print("Write to csv")
list_.to_csv("liste-komplett-{}.csv".format(time.strftime("%Y%m%d-%H%M%S")))

list_.head()
```
