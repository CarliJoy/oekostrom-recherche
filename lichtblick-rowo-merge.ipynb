# Arbeitslisten

Wir mergen die Liste, die Lichtblick uns freundlicherweise gegeben hat mit der bisherigen RoWo Rechercheliste.

```python
lichtblick = pd.read_csv("lichtblick/Lichtblick_Atromanbieter_Tabelle-clean.csv")
lichtblick['Lichtblick Liste'] = True
lichtblick.rename(columns={"Anbieter": "Firmenname"}, inplace=True)
lichtblick.head()
```

```python
rowo_list = pd.read_csv("lichtblick/RoWo-Arbeitsdok-2019-19-08-clean.csv")
rowo_list['RoWo Liste'] = True
rowo_list.rename(columns={"Stromanbieter": "Firmenname"}, inplace=True)
rowo_list.head()
```

```python
merged = pd.merge(lichtblick,
                 rowo_list,
                 on="Firmenname",
                 how="outer")

print(lichtblick.shape)
print(rowo_list.shape)
print(merged.shape)
merged.head()

merged.to_csv("lichtblick/merged-{}.csv".format(time.strftime("%Y%m%d-%H%M%S")))
```
